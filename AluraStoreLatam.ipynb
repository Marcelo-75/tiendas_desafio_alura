### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Unir todos los DataFrames en uno solo
import pandas as pd

df = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
```

```python
# Crear la columna de facturación
df['Facturacion'] = df['Precio'] + df['Costo de envío']

# Facturación total general
facturacion_total = df['Facturacion'].sum()

# Facturación promedio general
facturacion_promedio = df['Facturacion'].mean()

# Facturación por vendedor
facturacion_por_vendedor = df.groupby('Vendedor')['Facturacion'].sum().sort_values(ascending=False)

# Facturación por fecha
df['Fecha de Compra'] = pd.to_datetime(df['Fecha de Compra'], dayfirst=True)
facturacion_por_fecha = df.groupby('Fecha de Compra')['Facturacion'].sum()

# Fecha con mayor facturación
fecha_max_facturacion = facturacion_por_fecha.idxmax()
monto_max_facturacion = facturacion_por_fecha.max()

# Mostrar resultados
print(f"Facturación total: {facturacion_total}")
print(f"Facturación promedio por venta: {facturacion_promedio}")
print("\nFacturación total por vendedor:")
print(facturacion_por_vendedor.reset_index().rename(columns={'Facturacion': 'Facturación total'}))

print(f"\nLa fecha con mayor facturación fue el {fecha_max_facturacion.date()} con un total de {monto_max_facturacion}")
```

# 2. Ventas por categoría

```python
ventas_categoria = df.groupby('Categoría del Producto')['Producto'].count().reset_index()
ventas_categoria.columns = ['Categoría', 'Cantidad de ventas']
ventas_categoria = ventas_categoria.sort_values(by='Cantidad de ventas', ascending=False)

print("Ventas por categoría:")
print(ventas_categoria)
```

# 3. Calificación promedio de la tienda


```python
# Calificación promedio general
calificacion_promedio_general = df['Calificación'].mean()

# Calificación promedio por tienda (vendedor)
calificacion_por_vendedor = df.groupby('Vendedor')['Calificación'].mean().reset_index()
calificacion_por_vendedor.columns = ['Vendedor', 'Calificación promedio']
calificacion_por_vendedor = calificacion_por_vendedor.sort_values(by='Calificación promedio', ascending=False)

# Calificación por ciudad
calificacion_por_ciudad = df.groupby('Lugar de Compra')['Calificación'].mean().reset_index()
calificacion_por_ciudad.columns = ['Ciudad', 'Calificación promedio']
calificacion_por_ciudad = calificacion_por_ciudad.sort_values(by='Calificación promedio', ascending=False)

print(f"Calificación promedio general: {calificacion_promedio_general:.2f}")
print("\nCalificación promedio por vendedor:")
print(calificacion_por_vendedor)
print("\nCalificación promedio por ciudad:")
print(calificacion_por_ciudad)
```

# 4. Productos más y menos vendidos

```python
# Conteo de ventas por producto
ventas_producto = df['Producto'].value_counts()

producto_mas_vendido = ventas_producto.idxmax()
cantidad_mas_vendido = ventas_producto.max()
producto_menos_vendido = ventas_producto.idxmin()
cantidad_menos_vendido = ventas_producto.min()

# Valor total vendido del producto más vendido
valor_total_mas_vendido = df[df['Producto'] == producto_mas_vendido]['Facturacion'].sum()
# Vendedor con más ventas del producto más vendido
vendedor_top_producto = df[df['Producto'] == producto_mas_vendido]['Vendedor'].value_counts().idxmax()

# Lo mismo para el menos vendido
valor_total_menos_vendido = df[df['Producto'] == producto_menos_vendido]['Facturacion'].sum()
vendedor_menos_producto = df[df['Producto'] == producto_menos_vendido]['Vendedor'].value_counts().idxmax()

print(f"Producto más vendido: {producto_mas_vendido}")
print(f"Cantidad vendida: {cantidad_mas_vendido}")
print(f"Valor total facturado por este producto: {valor_total_mas_vendido}")
print(f"Vendedor con más ventas de este producto: {vendedor_top_producto}")

print("\nProducto menos vendido: ", producto_menos_vendido)
print(f"Cantidad vendida: {cantidad_menos_vendido}")
print(f"Valor total facturado por este producto: {valor_total_menos_vendido}")
print(f"Vendedor con más ventas de este producto: {vendedor_menos_producto}")
```

# 5. Envío promedio por tienda

```python
envio_promedio_vendedor = df.groupby('Vendedor')['Costo de envío'].mean().reset_index()
envio_promedio_vendedor.columns = ['Vendedor', 'Costo de envío promedio']
envio_promedio_vendedor = envio_promedio_vendedor.sort_values(by='Costo de envío promedio', ascending=False)

print("Costo de envío promedio por vendedor:")
print(envio_promedio_vendedor)
```

# 6. Informe Final

```python
tienda1 = pd.read_csv(url)
tienda1["Tienda"] = "Tienda 1"

tienda2 = pd.read_csv(url2)
tienda2["Tienda"] = "Tienda 2"

tienda3 = pd.read_csv(url3)
tienda3["Tienda"] = "Tienda 3"

tienda4 = pd.read_csv(url4)
tienda4["Tienda"] = "Tienda 4"

# Unimos todos los datos en un solo DataFrame
df = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)
```

```python
# Agregar columna de facturación
df['Facturacion'] = df['Precio'] + df['Costo de envío']

# Agrupamos por Tienda real (no por vendedor)
resumen_tiendas = df.groupby('Tienda').agg({
    'Precio': 'sum',
    'Costo de envío': 'sum',
    'Facturacion': 'sum',
    'Producto': 'count',
    'Calificación': 'mean'
}).reset_index()

resumen_tiendas.columns = ['Tienda', 'Total en ventas', 'Total en envíos', 'Facturación total', 'Cantidad de ventas', 'Calificación promedio']
resumen_tiendas['Lucro estimado'] = resumen_tiendas['Total en ventas']

# Ordenamos por menor lucro
resumen_ordenado = resumen_tiendas.sort_values(by='Lucro estimado')

# Seleccionamos la tienda con menos lucro
peor_tienda = resumen_ordenado.iloc[0]

print("🔍 Informe de evaluación por tienda\n")
print("Resumen general por tienda:")
resumen_ordenado_format = resumen_ordenado.copy()
cols_a_formatear = ['Total en ventas', 'Total en envíos', 'Facturación total', 'Cantidad de ventas', 'Lucro estimado']

for col in cols_a_formatear:
    resumen_ordenado_format[col] = resumen_ordenado_format[col].apply(lambda x: f"${x:,.0f}")

resumen_ordenado_format['Calificación promedio'] = resumen_ordenado_format['Calificación promedio'].round(2)

display(resumen_ordenado_format)

print("\n📉 Tienda con menor lucro estimado:")
print(f"- {peor_tienda['Tienda']}")
print(f"- Lucro estimado: ${peor_tienda['Lucro estimado']:,.0f}")
print(f"- Cantidad de ventas: {int(peor_tienda['Cantidad de ventas'])}")
print(f"- Calificación promedio: {peor_tienda['Calificación promedio']:.2f}")
print(f"- Facturación total (incluyendo envíos): ${peor_tienda['Facturación total']:,.0f}")

print("\n📌 Recomendación:")
print(f"La {peor_tienda['Tienda']} es la menos rentable según las métricas de ventas, lucro y calificación.")
print("Se sugiere evaluar si esta tienda debe mantenerse operativa, mejorar su estrategia comercial o considerar su cierre.")
```
