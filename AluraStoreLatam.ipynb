### Importaci贸n de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. An谩lisis de facturaci贸n



```python
# Unir todos los DataFrames en uno solo
import pandas as pd

df = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
```

```python
# Crear la columna de facturaci贸n
df['Facturacion'] = df['Precio'] + df['Costo de env铆o']

# Facturaci贸n total general
facturacion_total = df['Facturacion'].sum()

# Facturaci贸n promedio general
facturacion_promedio = df['Facturacion'].mean()

# Facturaci贸n por vendedor
facturacion_por_vendedor = df.groupby('Vendedor')['Facturacion'].sum().sort_values(ascending=False)

# Facturaci贸n por fecha
df['Fecha de Compra'] = pd.to_datetime(df['Fecha de Compra'], dayfirst=True)
facturacion_por_fecha = df.groupby('Fecha de Compra')['Facturacion'].sum()

# Fecha con mayor facturaci贸n
fecha_max_facturacion = facturacion_por_fecha.idxmax()
monto_max_facturacion = facturacion_por_fecha.max()

# Mostrar resultados
print(f"Facturaci贸n total: {facturacion_total}")
print(f"Facturaci贸n promedio por venta: {facturacion_promedio}")
print("\nFacturaci贸n total por vendedor:")
print(facturacion_por_vendedor.reset_index().rename(columns={'Facturacion': 'Facturaci贸n total'}))

print(f"\nLa fecha con mayor facturaci贸n fue el {fecha_max_facturacion.date()} con un total de {monto_max_facturacion}")
```

# 2. Ventas por categor铆a

```python
ventas_categoria = df.groupby('Categor铆a del Producto')['Producto'].count().reset_index()
ventas_categoria.columns = ['Categor铆a', 'Cantidad de ventas']
ventas_categoria = ventas_categoria.sort_values(by='Cantidad de ventas', ascending=False)

print("Ventas por categor铆a:")
print(ventas_categoria)
```

# 3. Calificaci贸n promedio de la tienda


```python
# Calificaci贸n promedio general
calificacion_promedio_general = df['Calificaci贸n'].mean()

# Calificaci贸n promedio por tienda (vendedor)
calificacion_por_vendedor = df.groupby('Vendedor')['Calificaci贸n'].mean().reset_index()
calificacion_por_vendedor.columns = ['Vendedor', 'Calificaci贸n promedio']
calificacion_por_vendedor = calificacion_por_vendedor.sort_values(by='Calificaci贸n promedio', ascending=False)

# Calificaci贸n por ciudad
calificacion_por_ciudad = df.groupby('Lugar de Compra')['Calificaci贸n'].mean().reset_index()
calificacion_por_ciudad.columns = ['Ciudad', 'Calificaci贸n promedio']
calificacion_por_ciudad = calificacion_por_ciudad.sort_values(by='Calificaci贸n promedio', ascending=False)

print(f"Calificaci贸n promedio general: {calificacion_promedio_general:.2f}")
print("\nCalificaci贸n promedio por vendedor:")
print(calificacion_por_vendedor)
print("\nCalificaci贸n promedio por ciudad:")
print(calificacion_por_ciudad)
```

# 4. Productos m谩s y menos vendidos

```python
# Conteo de ventas por producto
ventas_producto = df['Producto'].value_counts()

producto_mas_vendido = ventas_producto.idxmax()
cantidad_mas_vendido = ventas_producto.max()
producto_menos_vendido = ventas_producto.idxmin()
cantidad_menos_vendido = ventas_producto.min()

# Valor total vendido del producto m谩s vendido
valor_total_mas_vendido = df[df['Producto'] == producto_mas_vendido]['Facturacion'].sum()
# Vendedor con m谩s ventas del producto m谩s vendido
vendedor_top_producto = df[df['Producto'] == producto_mas_vendido]['Vendedor'].value_counts().idxmax()

# Lo mismo para el menos vendido
valor_total_menos_vendido = df[df['Producto'] == producto_menos_vendido]['Facturacion'].sum()
vendedor_menos_producto = df[df['Producto'] == producto_menos_vendido]['Vendedor'].value_counts().idxmax()

print(f"Producto m谩s vendido: {producto_mas_vendido}")
print(f"Cantidad vendida: {cantidad_mas_vendido}")
print(f"Valor total facturado por este producto: {valor_total_mas_vendido}")
print(f"Vendedor con m谩s ventas de este producto: {vendedor_top_producto}")

print("\nProducto menos vendido: ", producto_menos_vendido)
print(f"Cantidad vendida: {cantidad_menos_vendido}")
print(f"Valor total facturado por este producto: {valor_total_menos_vendido}")
print(f"Vendedor con m谩s ventas de este producto: {vendedor_menos_producto}")
```

# 5. Env铆o promedio por tienda

```python
envio_promedio_vendedor = df.groupby('Vendedor')['Costo de env铆o'].mean().reset_index()
envio_promedio_vendedor.columns = ['Vendedor', 'Costo de env铆o promedio']
envio_promedio_vendedor = envio_promedio_vendedor.sort_values(by='Costo de env铆o promedio', ascending=False)

print("Costo de env铆o promedio por vendedor:")
print(envio_promedio_vendedor)
```

# 6. Informe Final

```python
tienda1 = pd.read_csv(url)
tienda1["Tienda"] = "Tienda 1"

tienda2 = pd.read_csv(url2)
tienda2["Tienda"] = "Tienda 2"

tienda3 = pd.read_csv(url3)
tienda3["Tienda"] = "Tienda 3"

tienda4 = pd.read_csv(url4)
tienda4["Tienda"] = "Tienda 4"

# Unimos todos los datos en un solo DataFrame
df = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)
```

```python
# Agregar columna de facturaci贸n
df['Facturacion'] = df['Precio'] + df['Costo de env铆o']

# Agrupamos por Tienda real (no por vendedor)
resumen_tiendas = df.groupby('Tienda').agg({
    'Precio': 'sum',
    'Costo de env铆o': 'sum',
    'Facturacion': 'sum',
    'Producto': 'count',
    'Calificaci贸n': 'mean'
}).reset_index()

resumen_tiendas.columns = ['Tienda', 'Total en ventas', 'Total en env铆os', 'Facturaci贸n total', 'Cantidad de ventas', 'Calificaci贸n promedio']
resumen_tiendas['Lucro estimado'] = resumen_tiendas['Total en ventas']

# Ordenamos por menor lucro
resumen_ordenado = resumen_tiendas.sort_values(by='Lucro estimado')

# Seleccionamos la tienda con menos lucro
peor_tienda = resumen_ordenado.iloc[0]

print(" Informe de evaluaci贸n por tienda\n")
print("Resumen general por tienda:")
resumen_ordenado_format = resumen_ordenado.copy()
cols_a_formatear = ['Total en ventas', 'Total en env铆os', 'Facturaci贸n total', 'Cantidad de ventas', 'Lucro estimado']

for col in cols_a_formatear:
    resumen_ordenado_format[col] = resumen_ordenado_format[col].apply(lambda x: f"${x:,.0f}")

resumen_ordenado_format['Calificaci贸n promedio'] = resumen_ordenado_format['Calificaci贸n promedio'].round(2)

display(resumen_ordenado_format)

print("\n Tienda con menor lucro estimado:")
print(f"- {peor_tienda['Tienda']}")
print(f"- Lucro estimado: ${peor_tienda['Lucro estimado']:,.0f}")
print(f"- Cantidad de ventas: {int(peor_tienda['Cantidad de ventas'])}")
print(f"- Calificaci贸n promedio: {peor_tienda['Calificaci贸n promedio']:.2f}")
print(f"- Facturaci贸n total (incluyendo env铆os): ${peor_tienda['Facturaci贸n total']:,.0f}")

print("\n Recomendaci贸n:")
print(f"La {peor_tienda['Tienda']} es la menos rentable seg煤n las m茅tricas de ventas, lucro y calificaci贸n.")
print("Se sugiere evaluar si esta tienda debe mantenerse operativa, mejorar su estrategia comercial o considerar su cierre.")
```
